- Piechart for Categorical Values

```python
def autopct_format(values, labels):
    def my_format(pct, all_values):
        total = sum(all_values)
        val = int(round(pct * total / 100.0))
        return f'{pct:.1f}% ({val:,d})' if pct >= 7.5 else ''
    return lambda pct: my_format(pct, values)

for col in categorical_columns:
    
    all_categories = df_plot[col].unique()
    colors = sns.color_palette('pastel', len(all_categories))
    category_color_map = dict(zip(all_categories, colors))
    category_list = list(category_color_map.keys())

    fig, ax = plt.subplots(1, 2, figsize=(12, 6), subplot_kw={'aspect': 'equal'})

    income_0_counts = income_0[col].value_counts()
    income_1_counts = income_1[col].value_counts()
    
    income_0_counts.plot.pie(
        ax=ax[0],
        colors=[category_color_map[category] for category in income_0_counts.index],
        autopct=autopct_format(income_0_counts, category_list),
        startangle=140,
        textprops={'fontsize': 7},
        wedgeprops={'edgecolor': 'black', 'linewidth': 1.5},
        radius=1.0
    )
    
    ax[0].set_title(f'Pie Chart for {col} - Income <= 50K')
    ax[0].set_ylabel('')

    income_1_counts.plot.pie(
        ax=ax[1],
        colors=[category_color_map[category] for category in income_1_counts.index],
        autopct=autopct_format(income_1_counts, category_list),
        startangle=140,
        textprops={'fontsize': 7},
        wedgeprops={'edgecolor': 'black', 'linewidth': 1.5},
        radius=1.0
    )
    
    ax[1].set_title(f'Pie Chart for {col} - Income > 50K')
    ax[1].set_ylabel('')

    plt.show()
```

# Preprocessing

- Perbaikan kolom hubungan Punya Anak

```python
filtered_train_anak = df_train[df_train['Hubungan'] == 'Punya Anak']
filtered_test_anak = df_test[df_test['Hubungan'] == 'Punya Anak']

value_counts_jumlah_anak = df_train['jumlah_anak'].value_counts()
mode_jum_anak = df_train['jumlah_anak'].value_counts().index[0]

if df_train['jumlah_anak'].value_counts().index[0] == 0:
    mode_jum_anak = df_train['jumlah_anak'].value_counts().index[1]

df_train.loc[(df_train['Hubungan'] == 'Punya Anak') & (df_train['jumlah_anak'] == 0), 'jumlah_anak'] = mode_jum_anak
df_test.loc[(df_test['Hubungan'] == 'Punya Anak') & (df_test['jumlah_anak'] == 0), 'jumlah_anak'] = mode_jum_anak
```

- pendapatan and pengeluaran -> Financial Gain

```python
df_train['Financial Gain'] = df_train['pendapatan'] - df_train['pengeluaran']
df_test['Financial Gain'] = df_test['pendapatan'] - df_test['pengeluaran']

df_train['Financial Status'] = np.where(df_train['pendapatan'] > 0, 'pendapatan',
                                        np.where(df_train['pengeluaran'] > 0, 'pengeluaran',
                                        'balance'))
df_test['Financial Status'] = np.where(df_test['pendapatan'] > 0, 'pendapatan',
                                       np.where(df_test['pengeluaran'] > 0, 'pengeluaran',
                                       'balance'))

df_train = df_train.drop(['pendapatan', 'pengeluaran'], axis=1)
df_test = df_test.drop(['pendapatan', 'pengeluaran'], axis=1)
```

Mengangani nilai '?' pada Asal Negara dan '0' pada Financial Gain

- Asal Negara

```python
df_train['Asal Negara'] = df_train['Asal Negara'].replace('?', np.nan)
df_test['Asal Negara'] = df_test['Asal Negara'].replace('?', np.nan)
```

- Financial Gain

```python
df_train['Financial Gain'] = df_train['Financial Gain'].replace(0, np.nan)
df_test['Financial Gain'] = df_test['Financial Gain'].replace(0, np.nan)
```

Menangani nilai NaN

- Mencari kolom yang memiliki nilai NaN

```python
print("Data Train:\n")
display(df_train.isnull().sum())

print("Data Test:\n")
display(df_test.isnull().sum())
```

- Fill NaN dengan mode bagi data kategori dan median bagi data knotinu

```python
# Data Kategori
df_train['Pekerjaan'] = df_train['Pekerjaan'].fillna(df_train['Pekerjaan'].mode()[0])
df_test['Pekerjaan'] = df_test['Pekerjaan'].fillna(df_train['Pekerjaan'].mode()[0])

df_train['Kelas Pekerjaan'] = df_train['Kelas Pekerjaan'].fillna(df_train['Kelas Pekerjaan'].mode()[0])
df_test['Kelas Pekerjaan'] = df_test['Kelas Pekerjaan'].fillna(df_train['Kelas Pekerjaan'].mode()[0])

df_train['Asal Negara'] = df_train['Asal Negara'].fillna(df_train['Asal Negara'].mode()[0])
df_test['Asal Negara'] = df_test['Asal Negara'].fillna(df_train['Asal Negara'].mode()[0])

# Data Kontinu
df_train['Financial Gain'] = df_train['Financial Gain'].fillna(df_train['Financial Gain'].median())
df_test['Financial Gain'] = df_test['Financial Gain'].fillna(df_train['Financial Gain'].median())
```

# Feature Engineering

Concatenate Fitur

- Pendidikan and Jenjang Pendidikan

```python
df_train['Education'] = df_train['Pendidikan'] + " " + df_train['Jenjang Pendidikan']
df_test['Education'] = df_test['Pendidikan'] + " " + df_test['Jenjang Pendidikan']

df_train = df_train.drop(['Pendidikan', 'Jenjang Pendidikan'], axis=1)
df_test = df_test.drop(['Pendidikan', 'Jenjang Pendidikan'], axis=1)
```

- Pekerjaan and Kelas Pekerjaan

```python
df_train['Job'] = df_train['Pekerjaan'] + " " + df_train['Kelas Pekerjaan']
df_test['Job'] = df_test['Pekerjaan'] + " " + df_test['Kelas Pekerjaan']

df_train = df_train.drop(['Pekerjaan', 'Kelas Pekerjaan'], axis=1)
df_test = df_test.drop(['Pekerjaan', 'Kelas Pekerjaan'], axis=1)
```

- Hubungan and Status

```python
df_train['Family'] = df_train['Hubungan'] + " " + df_train['Status']
df_test['Family'] = df_test['Hubungan'] + " " + df_test['Status']

df_train = df_train.drop(['Hubungan', 'Status'], axis=1)
df_test = df_test.drop(['Hubungan', 'Status'], axis=1)
```

Labelling

```python
y = df_train['income'].replace({'>50K': 1, '<=50K': 0})
df_train = df_train.drop(['income'], axis=1)
```

# EDA 2

```python
from sklearn.feature_selection import SelectKBest, mutual_info_classif
from sklearn.model_selection import train_test_split
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
from imblearn.over_sampling import SMOTE
```

```python
X = df_train.copy()
y = pd.Series(y).astype('category')
X_sub = df_test.copy()
```

Correlation Bars

```python
mutual_info = mutual_info_classif(X, y, random_state=21)

mutual_info = pd.Series(mutual_info)
mutual_info.index = X.columns
mutual_info = pd.DataFrame(mutual_info.sort_values(ascending=False), columns=["Numerical_Feature_MI"])
mutual_info.style.background_gradient("cool")
```

- Standar Scaling

```python
def visualize_distribution(df, title):
    plt.figure(figsize=(10, 6))
    for col in df.columns:
        sns.kdeplot(df[col], fill=True, alpha=0.3)
    
    plt.title(title, fontsize=16)
    plt.xlabel('Value')
    plt.ylabel('Density')
    plt.show()
```

```python
scl = StandardScaler()
X_scaled = pd.DataFrame(scl.fit_transform(X), columns=X.columns)
X_sub_scaled = scl.transform(X_sub)
```

```python
display(
    visualize_distribution(X, 'Original Data Distribution'),
    visualize_distribution(X_scaled, 'Scaled Data Distribution')
)
```

- SMOTE, dan Split

```python
display("Sebelum SMOTE", y.value_counts())

smt = SMOTE(random_state=21)
X_smote, y_smote = smt.fit_resample(X_scaled, y)

display("Sesudah SMOTE", y_smote.value_counts())
```

```python
X_train, X_val, y_train, y_val = train_test_split(X_smote, y_smote, stratify=y_smote, random_state=21)
```

```python
(X_train.shape[0], X_val.shape[0])
```

```python
display(
    y_train.value_counts(),
    y_val.value_counts()
)
```

- Principal Component Analysis (PCA)

```python
# Assuming X_smote and y_smote are already defined
pca = PCA(n_components=2)
principal_components = pca.fit_transform(X_smote)

# Create DataFrame with principal components and labels
df_pca = pd.DataFrame(data=principal_components, columns=['PC1', 'PC2'])
df_pca['label'] = y_smote

# Adding more jitter
jitter_strength = 0.05  # Increase jitter strength
df_pca['PC1'] += np.random.normal(0, jitter_strength, size=df_pca['PC1'].shape)
df_pca['PC2'] += np.random.normal(0, jitter_strength, size=df_pca['PC2'].shape)

# Create scatter plot
plt.figure(figsize=(10, 8))  # Increase figure size for better visibility
sns.scatterplot(
    x='PC1', y='PC2',
    hue='label',
    palette='coolwarm',
    data=df_pca,
    s=50,  # Slightly increase point size for better visibility
    alpha=0.7,  # Decrease alpha for more transparency
    edgecolor='k', linewidth=0.5, marker='o'
)

plt.title('PCA of Binary Data with Jitter', fontsize=16)
plt.xlabel('Principal Component 1', fontsize=14)
plt.ylabel('Principal Component 2', fontsize=14)
plt.grid(True, linestyle='--', alpha=0.5)
plt.legend(title='Labels', fontsize=10)  # Add a legend for better understanding
plt.tight_layout()  # Adjust layout
plt.show()
```
